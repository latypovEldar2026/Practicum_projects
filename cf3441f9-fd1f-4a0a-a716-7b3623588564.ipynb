## Шаг 2. Инжиниринг признаков

### 2.1. Группы по срокам финансирования

Разделим все компании на три группы:

* Единичное финансирование — был всего один раунд финансирования.

* Срок финансирования до года — между первым и последним раундом финансирования прошло не более года.

* Срок финансирования более года.

Визуализируем соотношение этих групп, создав два графика:

* По количеству компаний: Покажите, какой процент от общего числа компаний относится к каждой из трёх групп.
* По объёму инвестиций: Отобразите, какую долю от общего объёма привлечённых средств получила каждая группа.

( Для ясности и согласованности используем единую цветовую палитру для всех графиков, чтобы каждая категория (например, «Единичное финансирование») всегда отображалась одним цветом)

```python
df['duration_days']=(df['last_funding_at']-df['first_funding_at']).dt.days
                     
def classify_company(row):
    if row['duration_days'] < 1:
        return 'Единичное финансирование'
    elif row['duration_days'] <= 365:
        return 'Срок финансирования до года'
    else:
        return 'Срок финансирования более года'

df['group_year'] = df.apply(classify_company, axis=1)
```

Оставим в столбце `market` только массовые сегменты. Для остальных сегментов заменим значения на заглушки — `niche` для нишевых и `mid` для средних.

Дальнейшие исследования выполнись с учётом этой замены. Индивидуальные сегменты внутри средней и нишевой групп не рассматривали  — они объединяются в два общих сегмента.


```python
def replace_market_value(segment):
    count = market_counts.get(segment, 0)
    if count > 120:
        return segment 
    elif 35 <= count <= 120:
        return 'mid'  
    else:
        return 'niche'  
    
df['market'] = df['market'].apply(replace_market_value)
```

Определим компании с аномальным объёмом общего финансирования — используем метод IQR отдельно по каждому сегменту(все нишевые сегменты должны быть объединены в одну группу, а средние — в другую)

Определим сегменты рынка с наибольшей долей компаний, получивших аномальное финансирование, и выведим топ таких сегментов.

```python
def find_anomalies_iqr(group):
    Q1 = group.quantile(0.25)
    Q3 = group.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    anomalies = (group < lower_bound) | (group > upper_bound)
    return anomalies

df['is_anomaly'] = df.groupby('market')['funding_total_usd'].transform(find_anomalies_iqr)

print(f'Всего компаниий с аномальным финансированием: {df["is_anomaly"].sum()}')

```

Всего компаниий с аномальным финансированием: 5202. Чтобы их различать от остальных, задали им критерий аномальности (столбец "is_anomaly")

```python

segment_analysis = df.groupby('market').agg(
    total_companies=('is_anomaly', 'count'),
    anomaly_companies=('is_anomaly', 'sum')
).reset_index()

segment_analysis['anomaly_ratio'] = (segment_analysis['anomaly_companies'] / segment_analysis['total_companies'])*100

top_segments = segment_analysis.sort_values('anomaly_ratio', ascending=False).head(10)
plt.figsize=(16, 18)
ax = top_segments.sort_values('anomaly_ratio', ascending=True).plot(kind='barh', x='market', y= 'anomaly_ratio')
ax.set_xlim(10, 20)
plt.title("Топ-10 сегментов с наибольшей долей компаний с аномальным финансированием")
plt.xlabel("Доля компаний с аномальным финансированием в %")
plt.ylabel("Сегменты")
plt.grid(axis='x')
plt.show()
```

Изучена структура сегментов и типов финансирования: проаналирована популярность и объёмы по типам (venture, seed, debt_financing и др.), оценена доля аномального финансирования по сегментам (максимум — 17% в Real Estate), остальные доли сегментов варируются от 16% до 16,5%,сопоставлены частота использования и объёмы для каждого типа.

### 4.3 Годовая динамика доли возвращённых средств по типам финансирования

Заказчик хочет знать, какая часть вложенных или выданных денег со временем возвращается обратно инвесторам или финансистам. Наша цель — для каждого года и каждого вида финансирования рассчитать нормированные значения возврата средств: то есть какую долю возвращённые средства составляют от предоставленных. При этом слишком большие аномальные значения, то есть неадекватные выбросы, нужно заменить на пропуски.

Совет: когда будете делить сумму возвращённых средств на суммарный объём привлечённого финансирования по конкретному году, добавьте к знаменателю небольшое число, например `1e-60`. Это поможет избежать деления на ноль.

```python
df['first_funding_at'] = pd.to_datetime(df['first_funding_at'])
df['funding_year'] = df['first_funding_at'].dt.year

funding_types = [
    'seed', 'venture', 'equity_crowdfunding', 'undisclosed',
    'convertible_note', 'debt_financing', 'angel', 'grant',
    'private_equity', 'post_ipo_equity', 'post_ipo_debt',
    'secondary_market', 'product_crowdfunding'
]

df_filtered = df[(df['funding_year'] >= 2000) & (df['funding_year'] <= 2014)]
result_table = df_filtered.groupby('funding_year')[funding_types].sum()
result_table = result_table[funding_types]
result_table.index = result_table.index.astype(int)
display(result_table)
```

2. Какой тип финансирования наиболее уместен?

Учитывая динамику возврата средств и объёмы инвестиций, рекомендуем:

Для стартапов и ранних проектов в E-Commerce, Technology, Social Network Media — венчурное финансирование (venture). Это наиболее популярный и успешный тип финансирования (более 40 000 компаний, объём свыше 12 млрд USD). Подходит для проектов с высоким потенциалом роста, но сопряжённых с рисками.

Для проектов с чётким денежным потоком (например, в сегменте E-Commerce с развитой логистикой) можно рассмотреть debt financing (заёмное финансирование), но с осторожностью из-за высокой волатильности.
Для посевных проектов (идеи на ранней стадии) — seed финансирование, однако следует учитывать относительно небольшие объёмы финансирования и высокие риски.
Ангельские инвестиции (angel) могут быть уместны для нишевых проектов с высоким потенциалом, но требуют индивидуального подхода.
